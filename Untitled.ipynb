```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import csv
```

```python
pos_value = pd.read_csv("posss.csv")
band_values = pd.read_csv("qwe.csv")
all_values = pd.read_csv("qwert.csv")
all = pd.read_csv("qwerty.csv")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
correlations = pos_value.corr()
cax = ax.matshow(correlations, vmin=0.0, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(["1-6","7-12","13-18","19-24"])
ax.set_yticklabels(["1-6","7-12","13-18","19-24"]);
```

```python
fig = plt.figure()
correlations = band_values.corr()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=0.0, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(["1-6","7-12","13-18","19-24"])
ax.set_yticklabels(["1-6","7-12","13-18","19-24"]);
```

```python
fig = plt.figure()
correlations = all_values.corr()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=0.6, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,8,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(["GK","DF","MF","FW","1-6","7-12","13-18","19-24"])
ax.set_yticklabels(["GK","DF","MF","FW","1-6","7-12","13-18","19-24"]);
```
