```python
import pandas as pd
from scipy.stats import shapiro

# Read CSV data
data = pd.read_csv("qwertyui.csv")

# Perform Shapiro-Wilk test on the "Value" column
shapiro_test_statistic, shapiro_p_value = shapiro(data["Value"])

# Format test statistic and p-value to three decimal places
formatted_test_statistic = "{:.3f}".format(shapiro_test_statistic)
formatted_p_value = "{:.3f}".format(shapiro_p_value)

print("Shapiro-Wilk Test Statistic:", formatted_test_statistic)
print("P-value:", formatted_p_value)

# Interpret the results
if shapiro_p_value < 0.05:
    print("The data is not normally distributed (reject null hypothesis)")
else:
    print("The data is normally distributed (fail to reject null hypothesis)")
```

```python
import pandas as pd
from scipy.stats import spearmanr

# Read CSV data
data = pd.read_csv("qwertyui.csv")

# List of unique League Standing values
league_standings = data["Position"].unique()

# Create an empty DataFrame to store correlation results
correlation_results = pd.DataFrame(columns=["Position", "Spearman's r results", "P-Value"])

# Perform Spearman's correlation for each League Standing category
for standing in league_standings:
    subset_data = data[data["Position"] == standing]
    correlation_coefficient, p_value = spearmanr(subset_data["Year"], subset_data["Value"])
    
    # Add results to the DataFrame
    correlation_results = correlation_results.append({
        "Position": standing,
        "Spearman's r results": correlation_coefficient,
        "P-Value": p_value
    }, ignore_index=True)

# Format correlation coefficients and p-values to three decimal places
correlation_results["Spearman's r results"] = correlation_results["Spearman's r results"].apply(lambda x: "{:.3f}".format(x))
correlation_results["P-Value"] = correlation_results["P-Value"].apply(lambda x: "{:.3f}".format(x))

# Display the correlation results DataFrame without the index column
correlation_results.style.hide_index()

dfi.export(correlation_results.style.hide_index(),"Spearman's R Position.png")
```
