## Graphs For Masters Dissertation

```python
#Import all packages needed for the graphs and calculations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv;
```

```python
#import all data as CSVs into the jupyter notebook
squad_value = pd.read_csv("Squad Value.csv")
avg_value = pd.read_csv("Avg Values.csv")
STD_Dev = pd.read_csv("STD Dev.csv");
```

```python
#Sets the size of the graph
plt.figure(figsize=(10,6))

#Plot first graph - bar graph with blue bars
plt.bar(squad_value['Year'], squad_value['Squad Value'])

#Plot second graph with orange points
plt.plot(squad_value['Year'], squad_value['Squad Value'], '-o', color='orange')

#Add x and y axis labels
plt.xlabel("Year", fontsize=11, fontname='Times New Roman')
plt.ylabel("Total Player Market Value in Millions", fontsize=11, fontname='Times New Roman')

#Add Graph Title
plt.title('Squad Value per Year', fontsize=12, fontname='Times New Roman')

#Make the y axis show every value
plt.xticks(squad_value["Year"])

#Save as png in folder
plt.savefig('Total Squad Value per Year');
```

```python
#Sets the size of the graph
plt.figure(figsize=(10,6))

#plot individual lines
plt.plot(STD_Dev["Year"], STD_Dev['Std Dev +1'], label='STD Dev +/-1',linestyle='dashed',color="blue")
plt.plot(STD_Dev["Year"],STD_Dev['Avg'], label='Avg Squad Value', linewidth=4)
plt.plot(STD_Dev["Year"],STD_Dev['Std Dev -1'],linestyle='dashed', color="blue")

#Name the labels and tilte, fint and fontsize
plt.ylabel('Average Squad Value in Millions', fontsize=11, fontname='Times New Roman')
plt.xlabel('Year', fontsize=11, fontname='Times New Roman')
plt.title('Squad Value with Std Deviations per Year', fontsize=12, fontname='Times New Roman')

#Make the y axis show every value
plt.xticks(STD_Dev["Year"])

#Add Legend
plt.legend()

#Save as png in folder
plt.savefig('Squad Value with Std Deviations per Year');
```

```python
#Sets the size of the graph
plt.figure(figsize=(10,6))

#State the order that the league standings will be organised in
hue_order = ["1 to 6","7 to 12","13 to 18","19 to 24"]

#Plot the graph, named plot so that titles and axis can be named later
plot = sns.lineplot(data=avg_value, x="Year", y="Average of Value in Millions", hue="Bands", hue_order=hue_order, errorbar=None)

#set the title on the y axis
plot.set_ylabel("Market Value in Millions", fontsize=11, fontname='Times New Roman')
plt.xlabel('Year', fontsize=11, fontname='Times New Roman')

#set the title for the graph
plot.set_title("Market Value in Millions versus Year by League Standing", fontsize=12, fontname='Times New Roman')

#set the legend title and plot legend
plt.legend(title="League Standing")

#Save as png in folder
plt.savefig('Market Value in Millions versus Year by League Standing');
```
